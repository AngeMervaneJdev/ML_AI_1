```python
import matplotlib.pyplot as plt
print("Probleme 1\n")
import numpy as np
cov = [[1.0,0.8],[0.8,1.0]]
mean = (-3,0)
np.random.seed(0)
A=np.array(np.random.multivariate_normal(mean,cov,500))

print(A)
```

```python
print('Probleme 2\n')
np.random.seed(0)
x,y=np.array(np.random.multivariate_normal(mean,cov,500).T)

plt.scatter(x,y)
plt.show()
```

```python
print("Probeme 3\n")
counts, bins = np.histogram(x,50)
plt.hist(bins[:-1], bins, weights=counts)
plt.xlim(left=-5.56,right=1.8) 
plt.show()
plt.clf()

counts, bins = np.histogram(y,50)
plt.xlim(left=-5,right=2.7) 
plt.hist(bins[:-1], bins, weights=counts)
plt.show()
```

```python
print("Probleme 4\n")
cov1 = [[1.0,0.8],[0.8,1.0]]
mean1 = (0,-3)
np.random.seed(0)
x1, y1 = np.random.multivariate_normal(mean1, cov1,500).T

plt.plot(x, y, 'x',color="yellow")
plt.plot(x1, y1, '*',color="blue")

plt.show()
```

```python
print("Probleme 5\n")
import numpy as np
cov1 = [[1.0,0.8],[0.8,1.0]]
mean1 = (0,-3)
np.random.seed(0)
B = np.array(np.random.multivariate_normal(mean1, cov1,500))

C=np.concatenate((A,B))
print(len(C))
```

```python
print("Probleme 6\n")
F=[]
for a in range(1000):
    if a<500:
        b=np.insert(C[a],2,0)
    else:
        b=np.insert(C[a],2,1)
    F.append(b)
print(len(F))
print(len(F[0]))
```
