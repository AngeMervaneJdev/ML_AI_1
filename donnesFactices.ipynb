```python
import matplotlib.pyplot as plt
print("Probleme 1\n")
import numpy as np
cov = [[1.0,0.8],[0.8,1.0]]
mean = (-3,0)
np.random.seed(0)
x, y = np.random.multivariate_normal(mean, cov,size=(500,2)).T
```

```python
print('Probleme 2\n')
plt.plot(x, y, 'x')
plt.show()
```

```python
print("Probeme 3\n")
counts, bins = np.histogram(x)
plt.hist(bins[:-1], bins, weights=counts)
plt.xlim(left=-6,right=2) 
plt.show()
plt.clf()

counts, bins = np.histogram(y)
plt.xlim(left=-3,right=2) 
plt.hist(bins[:-1], bins, weights=counts)
plt.show()
```

```python
print("Probleme 4\n")
cov1 = [[1.0,0.8],[0.8,1.0]]
mean1 = (0,-3)
np.random.seed(0)
x1, y1 = np.random.multivariate_normal(mean1, cov1,size=(500,2)).T

plt.plot(x, y, 'x',color="yellow")
plt.plot(x1, y1, '*',color="blue")

plt.show()
```
