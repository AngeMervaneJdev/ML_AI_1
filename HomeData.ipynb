Etude de mots

1- kurtosis

  le kurtosis est le coefficient d’acuité, coefficient d’aplatissement et degré de voussure, est une mesure directe de l’acuité et une mesure indirecte de l'aplatissement de la distribution d’une variable aléatoire réelle.

   Il mesure, abstraction faite de la dispersion (donnée par l’écart type), la répartition des masses de probabilité autour de leur centre, donné par l’espérance mathématique, c’est-à-dire, d’une certaine façon, leur concentration à proximité ou à distance du centre de probabilité.

   image.png

   
2- skewness

   est le coefficient d'asymétrie (skewness en anglais) correspond à une mesure de l’asymétrie de la distribution d’une variable aléatoire réelle.
   C’est le premier des paramètres de forme, avec le kurtosis (les paramètres basés sur les moments d’ordre 5 et plus n’ont pas de nom attribué).

En termes généraux, l’asymétrie d’une distribution est positive si la queue de droite (à valeurs hautes) est plus longue ou grosse, et négative si la queue de gauche (à valeurs basses) est plus longue ou grosse. 
image.png

```python
import seaborn as sns


print("\n Avant  transformation Logarithmique\n\n")
sns.histplot(data.SalePrice)
sns.displot(data.SalePrice)

# calcul du kurtosis

print("\n\n Kurtosis")
print(data.SalePrice.kurtosis())
print("\n\n Skewness")
print(data.SalePrice.skew())
```

```python

print("\n\nApres transformation Logarithmique\n\n")

log_salePrice=np.log(data.SalePrice)  # calcul du logarithme de la variable cible
print(log_salePrice)

print("\n\n Kurtosis")
print(log_salePrice.kurtosis())
print("\n\n Skewness")
print(log_salePrice.skew())

sns.histplot(log_salePrice)
sns.displot(log_salePrice)
```
