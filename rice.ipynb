```python
import numpy as np
import matplotlib.pyplot as plt

print("Probleme 1\n")


list_n_grains=[]
list_total_grains=[]

#the second to the 100e day
for day in range(1,101):
  list_n_grains.append(2**(day-1))
  list_total_grains.append(np.array(list_n_grains).sum())
  day+=1

print(list_total_grains)
print(list_n_grains)

plt.ylabel("number of rice grains you receive on that day ")
plt.xlabel("number of days")
plt.plot(list_n_grains)
plt.show()
plt.clf()

plt.ylabel("number of total rice grains you receive on that day ")
plt.xlabel("number of days")
plt.plot(list_total_grains)
plt.show()

```

```python
print("Probleme 2\n")
import numpy as np
def compute_sorori_shinzaemon(n_days=100):
 
    n_days: int
   
    list_n_grains : list
      
    list_total_grains : list

    # The first day
    quantity=1
    list_total_grains=[]
    list_n_grains=[]
    list_total_grains.append(1)
    list_n_grains.append(1)

    #the second to the 100e day
    while n_days>1:
      quantity=quantity*2
      list_n_grains.append(quantity)
      list_total_grains.append(np.array(list_n_grains).sum())
      n_days-=1
    
    return list_n_grains, list_total_grains

print("For 100 days we have\n")
list_n_grains, list_total_grains = compute_sorori_shinzaemon(n_days=100)
print("n rice receive the 100e day {}".format(list_n_grains[-1]))
print("n total nrice receive the 100e day {}\n".format(list_total_grains[-1]))
print("----------------------\n")
print("For exemple, 123 days we have\n")
list_n_grains, list_total_grains = compute_sorori_shinzaemon(n_days=123)
print("n rice receive the 123e day {}".format(list_n_grains[-1]))
print("n total nrice receive the 123e day {}".format(list_total_grains[-1]))
```
