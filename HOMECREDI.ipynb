```python
print("Probleme 3\n")
import numpy as np
import pandas as pd
data=pd.read_csv("application_train.csv")
import matplotlib.pyplot as plt
import seaborn as sns

# question 1
# age moyen des clients
plt.clf()
print("Age moyen des clients "+str(int(-(data.DAYS_BIRTH/365).sum()/len(data.DAYS_BIRTH)))+" ans")
print((data.DAYS_BIRTH/365).index)
plt.plot(data.DAYS_BIRTH/365)
plt.show()

# question 2
print("La corrélation entre AMT_INCOME_TOTAL et CNT_CHILDREN est "+str(data.AMT_INCOME_TOTAL.corr(data.CNT_CHILDREN)))
sns.heatmap(data[["AMT_INCOME_TOTAL","CNT_CHILDREN"]].corr(),linewidths=0.4,cbar_kws={"shrink":.75},vmax=1,vmin=-1,annot=True)

plt.show()
# question 4
print("Les types d'occupatiobs des clients \n")
print(data.OCCUPATION_TYPE)
plt.clf()

print("La corrélation entre le nombre de membre de la famille (CNT_FAM_MEMBERS) et le remboursement est 0.0023")
sns.heatmap(data[["CNT_FAM_MEMBERS","TARGET"]].corr(),linewidths=0.4,cbar_kws={"shrink":.75},vmax=1,vmin=-1,annot=True)
plt.show()
plt.clf()
print('Les données en corrélation avec le TARGET')
plt.figure(figsize=(15,10))

# question 8   Tableau de corrélation avec le TARGET

corr_with_target=data.corrwith(data.TARGET,drop=False)
corr_10_data=corr_with_target.sort_values(ascending=False)
sns.heatmap(data[corr_10_data.index].corr(),linewidths=0.8,cbar_kws={"shrink":.75},vmax=1,vmin=-1,annot=False)
plt.show()
plt.clf()
# question 9

corr_10_data=corr_10_data[1:11] # 

print(pd.DataFrame(data=corr_10_data,columns=["correlation"]))

sns.heatmap(data[corr_10_data.index].corr(),linewidths=0.4,cbar_kws={"shrink":.75},vmax=1,vmin=-1,annot=True)
plt.show()
plt.clf()
```
