```python
print('Probleme 1\n')
import pandas as pd
data=pd.read_csv("train.csv")
X=data.GrLivArea
Y=data.YearBuilt
Z=data.SalePrice
print(X,Y,Z)
```

```python
print("Probleme 2\n")
# import Sk Learn train test split
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = data[["GrLivArea","YearBuilt"]]# varible descriptive
y = data["SalePrice"]  # variable explicative

# Assign variables to capture train test split output
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Pretraitement et standardisation

scaler = StandardScaler()
scaler.fit(X_train)

# Pretraitement et standardisation

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

from sklearn.linear_model import LinearRegression

#instanciation et ajustement du modele aux donees d'entrainement
lreg = LinearRegression()

#Fit model to training data
lreg.fit(X_train, y_train)

# integration / * *** * /

# /* ***** */
y_pred = lreg.predict(X_test)

print("Y predicted "+str(y_pred.shape))

print("X test "+str(X_test.shape))

# vizualisation
import matplotlib.pyplot as plt

###GrLivArea

plt.subplots(figsize=(15, 7)) 
plt.xlabel('GrLivArea')
plt.ylabel('SalePrice')
plt.title('GrLivArea predicted')
plt.scatter(X_test[:,0], y_pred, c="blue")
plt.show()
## YearBuilt
plt.clf()
plt.subplots(figsize=(15, 7)) 
plt.xlabel('YearBuilt')
plt.ylabel('SalePrice')
plt.title('YearBuilt predicted')
plt.scatter(X_test[:,1], y_pred, c="red")
plt.show()
```

```python
print('Probleme 3\n')
from sklearn.model_selection import cross_val_score

def show_graph(y_pred,model):
  print()
  print("*"*10)
  print("\nCross_val_score  mean : "+str(cross_val_score(model, X_test, y_pred, cv=10).mean())+"\n\n")
  plt.clf()
  plt.subplots(figsize=(15, 7)) 
  plt.xlabel('YearBuilt')
  plt.ylabel('SalePrice')
  plt.title('YearBuilt predicted')
  plt.scatter(X_test[:,0], y_pred, c="blue")
  plt.show()

  plt.clf()
  plt.subplots(figsize=(15, 7)) 
  plt.xlabel('GrLivArea')
  plt.ylabel('SalePrice')
  plt.title('GrLivArea predicted')
  plt.scatter(X_test[:,1], y_pred, c="red")
  plt.show()


# Model 1 LinearRegression
print('# Model 1 LinearRegression')
lreg = LinearRegression().fit(X_train, y_train)
y_pred_lreg = lreg.predict(X_test)
show_graph(y_pred_lreg,lreg)

# Model 2 SVM
print("# Model 2 SVM")
from sklearn.svm import SVR
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
regr = make_pipeline(StandardScaler(), SVR(C=1.0, epsilon=0.2))
regr.fit(X_train, y_train)
y_pred_svm=regr.predict(X_test)
show_graph(y_pred_svm,regr)

# Model 3 DecisionTree
print("# Model 3 DecisionTree")
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(X_train, y_train)
y_pred_DT=regressor.predict(X_test)

show_graph(y_pred_DT,regressor)

# Model 4 RandomForest
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression
regr4 = RandomForestRegressor(max_depth=2, random_state=0)
regr4.fit(X_train, y_train)
y_pred_RF=regr4.predict(X_test)
show_graph(y_pred_RF,regr4)
```
